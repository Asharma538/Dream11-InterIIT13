```python
import os
import time
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import google.generativeai as genai
```

```python
df = pd.read_csv("names_csvs/players_with_cricinfo_id.csv")
```

```python
df
```

```python
df['full name'] = ""
df['playing role'] = ""
```

```python
for i in range(len(df["id_cricinfo"])):
    pid = df["id_cricinfo"][i]
    res = requests.get(f"https://www.espncricinfo.com/cricketers/x-"+str(pid)).text

    soup = BeautifulSoup(res, 'html.parser')
    parent_element = soup.find(class_="ds-grid")
    if parent_element:
        for child in parent_element.find_all(recursive=False):
            first_child_text = child.contents[0].text.strip() if child.contents else ""
            second_child_text = child.contents[1].text.strip() if len(child.contents) > 1 else ""
            # Example of how to fill the new columns with data
            if first_child_text.lower() == "full name":
                df.at[i, 'full name'] = second_child_text
                print(second_child_text)
            if first_child_text.lower() == "playing role":
                df.at[i, 'playing role'] = second_child_text
    
```
